```python
import numpy as np
```

```python
number_of_points = 500

x_point = []
y_point = []

a = 0.22
b = 0.78
```

```python
for i in range(number_of_points):
    x = np.random.normal(0., 0.5)
    y = a*x + b + np.random.normal(0., 0.1)
    x_point.append([x])
    y_point.append([y])
```

```python
import matplotlib.pyplot as plt

plt.plot(x_point, y_point, 'o', label = 'Input Data')
plt.legend()
plt.show()
```

```python
# not using for loop
# maybe better than example on the book?
```

```python
x_v_point = np.random.normal(0., 0.5, size=number_of_points)
y_v_point = a * x_v_point + np.random.normal(0., 0.1, size=number_of_points)
plt.plot(x_v_point, y_v_point, 'o', label = 'Input Data')
plt.legend()
plt.show()
```

```python
import tensorflow as tf

A = tf.Variable(tf.random_uniform(shape = [1], minval=-1., maxval=1.))
b = tf.Variable(tf.zeros([1]))

y = A * x_v_point + b
```

```python
# mse
cost_function = tf.reduce_mean(tf.square(y - y_v_point))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.5)
```

```python
train = optimizer.minimize(cost_function)
```

```python
model = tf.global_variables_initializer()#tf.initialize_all_variables()
```

```python
with tf.Session() as session:
    session.run(model)
    for step in range(0,21):
        session.run(train)
        if (step%5)==0:
            plt.plot(x_v_point, y_v_point, 'o', label = 'step = {}'.format(step))
            plt.plot(x_v_point, session.run(A)*x_v_point + session.run(b))
            plt.legend()
            plt.show()
```
